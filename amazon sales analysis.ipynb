# AMAZON SALES REPORT ANALYSIS

# Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the CSV file:

```python
file_path = 'Downloads/Amazon Sale Report.csv'
sales_data = pd.read_csv(file_path)
```

# Convert 'Date' to datetime format

```python
sales_data['Date'] = pd.to_datetime(sales_data['Date'], errors='coerce')
```

# Calculate total sales over time

```python
sales_data.set_index('Date', inplace=True)
sales_over_time = sales_data['Amount'].resample('M').sum()
sales_over_time.plot(kind='line', title='Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

# 2.Product Analysis

 Objective: Analyze the distribution of product categories, sizes, and quantities sold to identify popular products.

# (A)Sales by Category:

```python
sales_by_category = sales_data.groupby('Category')['Amount'].sum().sort_values(ascending=False)
print(sales_by_category)
```

```python
plt.figure(figsize=(12, 6))
sales_by_category.plot(kind='bar', title='Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.savefig('sales_by_category.png')
plt.show()
```

# (B)Sales by Size:

```python
sales_by_size = sales_data.groupby('Size')['Amount'].sum().sort_values(ascending=False)
print(sales_by_size)
```

```python
plt.figure(figsize=(12, 6))
sales_by_size.plot(kind='bar', title='Sales by Size')
plt.xlabel('Size')
plt.ylabel('Total Sales')
plt.savefig('sales_by_size.png')
plt.show()
```

# 3. Fulfillment Analysis

Objective: Investigate the fulfillment methods used and their effectiveness in delivering orders.

# (A)Sales by Fulfillment Method:

```python
sales_by_fulfillment = sales_data.groupby('Fulfilment')['Amount'].sum().sort_values(ascending=False)
print(sales_by_fulfillment)
```

```python
plt.figure(figsize=(12, 6))
sales_by_fulfillment.plot(kind='bar', title='Sales by Fulfillment Method')
plt.xlabel('Fulfillment Method')
plt.ylabel('Total Sales')
plt.savefig('sales_by_fulfillment.png')
plt.show()
```

# (B)Order Status Distribution:

```python
order_status_distribution = sales_data['Status'].value_counts()
print(order_status_distribution)
```

```python
plt.figure(figsize=(12, 6))
order_status_distribution.plot(kind='bar', title='Order Status Distribution')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.savefig('order_status_distribution.png')
plt.show()
```

# 4. Customer Segmentation

Objective: Segment customers based on their buying behavior, location, and other relevant factors.

# (A)Segment by Location:

```python
sales_by_city = sales_data.groupby('ship-city')['Amount'].sum().sort_values(ascending=False)
print(sales_by_city)
```

```python
plt.figure(figsize=(12, 6))
sales_by_city.head(20).plot(kind='bar', title='Top 20 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.savefig('sales_by_city.png')
plt.show()
```

# (B)Segment by Buying Behavior (e.g., repeat customers):



```python
repeat_customers = sales_data['Order ID'].value_counts()
print(repeat_customers[repeat_customers > 1])
```

# 5. Geographical Analysis

Objective: Explore the geographical distribution of sales, focusing on states and cities.

# (A)Sales by State:

```python
sales_by_state = sales_data.groupby('ship-state')['Amount'].sum().sort_values(ascending=False)
print(sales_by_state)
```

```python
plt.figure(figsize=(12, 6))
sales_by_state.plot(kind='bar', title='Sales by State')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.savefig('sales_by_state.png')
plt.show()
```
